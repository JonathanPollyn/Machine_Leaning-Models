```python
# Importing the required libraries
import pandas as pd 
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Loading the data
filePath = 'data/bigmart.csv'
data = pd.read_csv(filePath)
data.head()
```

## Performing Square root on the data 

```python
fig = plt.figure()
sns.distplot(data['Item_Outlet_Sales'], fit = norm)

fig = plt.figure()
res_plt = stats.probplot(data['Item_Outlet_Sales'], plot = plt)
plt.show()
```

```python
# Applying the square root ont he data
data['Item_Outlet_Sales'] = np.sqrt(data['Item_Outlet_Sales'])
```

```python
fig = plt.figure()
sns.distplot(data['Item_Outlet_Sales'], fit = norm)

fig = plt.figure()
res_plt = stats.probplot(data['Item_Outlet_Sales'], plot = plt)
plt.show()
```

## Performing cube root on the data 

```python
# Loading the data
filePath = 'data/bigmart.csv'
data = pd.read_csv(filePath)
data.head()
```
