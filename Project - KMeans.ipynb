## Data Preparation for model descovery

```python
data.isna().sum()
```

```python
# Countries that have Nan in Agriculture
data[data['Agriculture'].isna()]['Country']
```

```python
fillNan = data[data['Agriculture'].isna()].fillna(0)
```

```python
data[data['Agriculture'].isna()] = fillNan
```

```python
data.isna().sum()
```

```python
data['Climate'] = data['Climate'].fillna(data.groupby('Region')['Climate'].transform('mean'))
```

```python
data.isna().sum()
```

```python
data['Literacy (%)'] = data['Literacy (%)'].fillna(data.groupby('Region')['Literacy (%)'].transform('mean'))
```

```python
data.isna().sum()
```

```python
data = data.dropna()
```

```python
data.isna().sum()
```

## Data Feature Preparation

```python
# data.head()
```

```python
X = data.drop(['Country'], axis=1)
```

```python
# X
```

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler_X = scaler.fit_transform(X)
```

```python
scaler_X
```

## Creating and Fitting the KMean Model

```python
from sklearn.cluster import KMeans
```

```python
ssd = []

for k in range(2,30):
    model = KMeans(n_clusters=k)
    model.fit(scaler_X)
    
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,30), ssd, 'o--')
plt.xlabel("K values")
plt.ylabel("Sum of squred distance")
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```

## Model Interpretation

```python
model = KMeans(n_clusters=3)
model.fit(scaler_X)
```

```python
model.labels_
```

```python
X['k=3 Clusters'] = model.labels_
```

```python
X.corr()['k=3 Clusters'].sort_values()
```
