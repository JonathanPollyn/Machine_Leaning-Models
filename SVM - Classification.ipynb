```python
# Importing the required libral
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
filePath = 'data/mouse_viral_study.csv'
data = pd.read_csv(filePath)
data.head()
```

```python
data.columns.to_list()
```

Creating Classes

```python
sns.scatterplot(x='Med_1_mL', y='Med_2_mL', data=data, hue='Virus Present')
```

Hyperplane

```python
sns.scatterplot(x='Med_1_mL', y='Med_2_mL', data=data, hue='Virus Present')

x = np.linspace(0,10,1000)
m = -1 
b = 11
y = m*x + b 
plt.plot(x,y)
```

SVM - Support Vector machine

```python
from sklearn.svm import SVC
```

```python
# help(SVC)
```

```python
X = data.drop(['Virus Present'], axis=1)
y = data['Virus Present']
```

```python
model = SVC(C=1000, kernel='linear')
model.fit(X, y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model, X, y)
```

Working with Kernel

```python
model = SVC(kernel='rbf', C=1)
model.fit(X,y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='sigmoid')
model.fit(X,y)
plot_svm_boundary(model, X, y)
```

Degree 

```python
model = SVC(kernel='poly', C=1, degree=1)
model.fit(X,y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='poly', C=1, degree=2)
model.fit(X,y)
plot_svm_boundary(model, X, y)
```

Gamma

```python
model = SVC(kernel='rbf', C=1, gamma=0.01)
model.fit(X,y)
plot_svm_boundary(model, X, y)
```

Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C': [0.01,0.1,1],'kernel':['linear','rbf']}
grid = GridSearchCV(svm, param_grid)
```

```python
grid.fit(X,y)
```

```python
grid.best_score_
```

```python
grid.best_params_
```
